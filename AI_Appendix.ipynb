### How I Used AI Tools

I used Claude (Anthropic) selectively throughout this project. My use fell into three categories:

**Boilerplate and repetitive writing:** sections that are structurally consistent but tedious to write from scratch, such as markdown cell headers, section objective blocks, and the formatting of this written report. Claude was used to generate an initial draft which I then revised to reflect my own analysis and wording.

**Plot enhancements:** I wrote the core visualisation logic myself. Claude helped add finishing details such as axis labels, titles, figure captions, and the written interpretation lines beneath each plot. All interpretation content was reviewed and confirmed against my own reading of the data before being kept.

**Report structuring:** Claude was used to help organise the Written Report into a clean, readable format. The findings, numbers, and conclusions are my own. Claude helped arrange them into a consistent structure and plain-language prose.
