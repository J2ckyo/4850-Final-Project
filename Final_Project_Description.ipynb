# Mini-Project: From Data Insight to Prediction

In this mini-project, you will work with a real-world dataset to move from raw data to insight and finally to
prediction. You will clean and explore a dataset, answer one meaningful statistical question, and then build a
simple predictive model. You will also present your work in class, focusing on clarity, reasoning, and
interpretation rather than code details. This project is designed to assess your ability to combine data wrangling,
visualization, statistics, modelling, and communication into a single, coherent analysis.

**1. Dataset Requirements**

You must choose one publicly available dataset. Your dataset must meet all of the following:

- At least 5,000 rows after cleaning
- At least 8 columns
- At least 4 numeric variables
- At least 2 categorical variables
- A clearly defined target variable that is:
    o numeric (for linear regression), or
    o binary (for logistic regression)

You may combine multiple datasets if there is a clear and sensible join. If you do this, you must explain how
and why the datasets were merged.

**2. Data Wrangling and Feature Engineering**

You must demonstrate at least four meaningful data-processing steps, such as:

- handling missing values
- fixing data types
- creating new features
- recoding or grouping categories
- filtering or reshaping data
- merging datasets

You must clearly state:

- the original shape of the dataset
- the final shape after cleaning
- why each wrangling step was necessary


**3. Exploratory Data Analysis**

Use visualizations and summary statistics to understand your data. Focus on insight, not volume. Every figure
should have a purpose.

Required:

- 2 to 4 well-chosen plots
- Each plot must include:
    o a title
    o labeled axes
    o a brief written interpretation (1 to 2 sentences)
- A short table of descriptive statistics
- A correlation analysis relevant to your target variable
**4. Statistical Question**

You must pose one clear statistical question and answer it using an appropriate statistical test.

Examples:

- Is there a significant difference between two groups?
- Is there a meaningful association between two variables?

You must include:

- the test used
- a brief justification for the test
- the test result (p-value)
- an effect size or confidence interval
- a plain-language conclusion
**5. Predictive Modelling**

You will build one regression model:

- Linear regression for a numeric target, or
- Logistic regression for a binary target

Model requirements:

- Split data into:
    o 60% training
    o 20% validation
    o 20% test
- Use a pipeline that includes:
    o scaling numeric variables
    o encoding categorical variables


```
o Tune at least one model parameter (where appropriate)
```
Report the following (only test-set results should be reported as final performance).

For regression:

- RMSE
- MAE

For classification:

- Accuracy
- Precision
- Recall
- F1 score
- ROC-AUC
**6. Interpretation and Insight**

This section matters as much as the code. Write clearly and concisely. Assume your audience is intelligent but
not technical. You must:

- interpret the most important model coefficients
- explain what the model suggests about the real-world problem
- identify at least one limitation
- propose one concrete next step or recommendation
**7. In-Class Presentation**

You will give a short in-class presentation summarizing your project. The focus is on communication, insight,
and reasoning, not technical detail.

Presentation guidelines:

- 5 minutes maximum
- Slides are recommended but not required
- Do not show code unless it is essential

Your presentation should cover:

- The dataset and research question
- One or two key insights from exploration or statistics
- The model used and how well it performed
- What the results mean in practical terms
- One limitation or next step


**8. Deliverables**

You must submit all of the following:

1. Jupyter Notebook
    - Runs from top to bottom without errors
    - Clear markdown explanations throughout
    - Clean, readable code
    - Figures saved to a folder
    - Tables saved to a folder
2. Written Report (PDF)
    - Maximum 5 pages
    - Includes:
       o project overview
       o key visuals
       o statistical results
       o model performance table
       o interpretation and conclusion
3. Dataset Files
    - All data files needed to reproduce your work
4. AI Use Appendix. This section is about transparency and good practice.

Include a short appendix describing:

- how you used AI tools (if at all)
- what types of help you asked for
- how you verified the results
**9. Evaluation Emphasis**

Strong projects tell a clear data story from start to finish, both in writing and in presentation.

Your work will be evaluated on:

- appropriateness and correctness of methods
- quality of interpretation and reasoning
- clarity of written and spoken communication
- coherence of the full workflow
- professionalism and reproducibility
